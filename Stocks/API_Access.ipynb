# Accessing the API directcly in Google Sheets 

```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials

# Use the JSON key file to authenticate
scope = ['https://spreadsheets.google.com/feeds', 'https://www.googleapis.com/auth/drive']
creds = ServiceAccountCredentials.from_json_keyfile_name('finanal-7bbc65b6c0b4.json', scope)
client = gspread.authorize(creds)

# Open the Google Sheet by title
# sheet = client.open("Calcul").sheet1
spreadsheet = client.open_by_key('https://docs.google.com/spreadsheets/d/1-DVqbqQSh5dXE02jOH4uTQejzhjMgvDk40wmZEQrxsU/edit#gid=419853359')





# Read data
data = sheet.get_all_records()
print(data)
```
